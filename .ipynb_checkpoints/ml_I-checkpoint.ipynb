```python
li=[i for i in range (1000000)]
start_time = process_time()
li = [i+5 for i in li]
end_time = process_time()
print(f"{end_time - start_time:.6f}")
```

```python
import numpy as np
from time import process_time

# Create a numpy array with 1 million elements
np_arr = np.array([i for i in range(1000000)])

# Measure the time it takes to add 5 to each element
start_time = process_time()
np_arr += 5  # Vectorized operation to add 5 to each element
end_time = process_time()

# Print the execution time
print(f"Execution time: {end_time - start_time:.6f} seconds")
```
